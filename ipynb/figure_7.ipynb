```r
library(ggplot2)
library("ggbeeswarm")
df <- read.table("../data/S5_Dataset.txt",header = TRUE)
```

```r
GPX1 <- unlist(as.list(df[c("ENSG00000233276"),]))
GPX2 <- unlist(as.list(df[c("ENSG00000176153"),]))
GPX3 <- unlist(as.list(df[c("ENSG00000211445"),]))
GPX4 <- unlist(as.list(df[c("ENSG00000167468"),]))
GPX5 <- unlist(as.list(df[c("ENSG00000224586"),]))
GPX6 <- unlist(as.list(df[c("ENSG00000198704"),]))
GPX7 <- unlist(as.list(df[c("ENSG00000116157"),]))
GPX8 <- unlist(as.list(df[c("ENSG00000164294"),]))
GPX <- data.frame( FPKM=c(GPX1,GPX2,GPX3,GPX4,GPX5,GPX6,GPX7,GPX8),
                  GPX=c(rep("GPX1",462),
                        rep("GPX2",462),
                        rep("GPX3",462),
                        rep("GPX4",462),
                        rep("GPX5",462),
                        rep("GPX6",462),
                        rep("GPX7",462),
                        rep("GPX8",462)))
```

# figure 7
Simple way of finding the contribution of GPX7 relative to all other GPX genes

```r
options(repr.plot.width=12, repr.plot.height=8)
lambda <-ggplot(GPX, aes(x=GPX, y=log(FPKM), fill=GPX, color=GPX))
lambda <- lambda + geom_quasirandom(data=GPX, alpha = .6, width=.3,show.legend = FALSE,size=.6)
lambda <- lambda + geom_boxplot(data=GPX,show.legend = FALSE,outlier.shape = NA,alpha = .3)
lambda <- lambda + theme_bw()
lambda <- lambda + scale_fill_manual(values=c('#EE6677','#EE6677','#EE6677','#EE6677','#EE6677','#EE6677','#4477AA','#EE6677'))
lambda <- lambda + scale_color_manual(values=c('#EE6677','#EE6677','#EE6677','#EE6677','#EE6677','#EE6677','#4477AA','#EE6677'))
lambda <- lambda + ggtitle(bquote("derivation of " ~ lambda))
lambda <- lambda +  ylab(bquote("ln(count)"))
lambda <- lambda + scale_y_continuous(limits = c(0,10))
lambda <- lambda + theme(text = element_text(size = 24))
lambda <- lambda + theme ( panel.background = element_rect (fill = "white"),axis.title.x=element_blank())
lambda <- lambda + theme(axis.text.x=element_text(angle=90, size=24, vjust=.5,hjust=1.0))
plot(lambda)
```
